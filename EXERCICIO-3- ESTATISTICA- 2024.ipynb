```python
df=pd.read_csv('DEMOGRAFIA_FORTALEZA.csv')
df=df.dropna(axis=1)
df=df.rename(columns={'Bairro.1':'BAIRRO'}).reset_index(drop=True)
df=df[df['Pop. Homens']!=0].reset_index(drop=True)
```

```python
df['total_pop'] = df['Pop. Homens'] + df['Pop. Mulheres']
```

# 2 Questao

```python
# População entre 0-14
plt.scatter(df['População (0-14 anos)'],df['total_pop'],edgecolors='black')
plt.xlabel('População (0-14 anos)')
plt.ylabel('total_pop')
plt.show()
```

```python
# M
x=df['População (0-14 anos)'].values
y=df['total_pop'].values
```

```python
tamanho=len(x)
```

```python
nteste=1000
mvalues=[]
for _ in range(nteste):
    i=np.random.randint(0,tamanho)
    j=np.random.randint(0,tamanho)
    if i!=j:
        m=(y[i]-y[j])/(x[i]-x[j])
        mvalues.append(m)
```

```python
plt.hist(mvalues,bins=50,edgecolor='black',color='blue')
plt.show()
```

```python
ser=pd.Series(mvalues)
```

```python
ser.describe()
```

```python
q1=ser.quantile(0.25)
q3=ser.quantile(0.75)
iqr=q3-q1
liminf=q1-1.5*iqr
limsup=q3+1.5*iqr
print(liminf,limsup)
```

```python
ser[(ser>=liminf)&(ser<=limsup)].describe()
```

```python
ser_filtro = ser[(ser>=liminf)&(ser<=limsup)]
```

```python
plt.hist(ser_filtro,bins=50,edgecolor='black',color='blue')
plt.axvline(ser_filtro.mean(),ls='--',color='red')
plt.axvline(ser_filtro.median(),ls='--',color='orange')
plt.show()
```

```python
m=ser_filtro.mean()
```

```python
m
```

```python
# N
nvalues=[]
for i in range(len(y)):
    n=y[i]-m*x[i]
    nvalues.append(n)
```

```python
nser=pd.Series(nvalues)
```

```python
plt.hist(nser,bins=50,edgecolor='black',color='blue')
plt.axvline(nser.mean(),ls='--',color='red')
plt.axvline(nser.median(),ls='--',color='orange')
plt.show()
```

```python
nser.describe()
```

```python
q1=nser.quantile(0.25)
q3=nser.quantile(0.75)
iqr=q3-q1
liminf=q1-1.5*iqr
limsup=q3+1.5*iqr
print(liminf,limsup)
```

```python
nser_filtro=nser[(nser>=liminf)&(nser<=limsup)]
```

```python
nser_filtro.describe()
```

```python
plt.hist(nser_filtro,bins=10,edgecolor='black',color='blue')
plt.axvline(nser_filtro.mean(),ls='--',color='red')
plt.axvline(nser_filtro.median(),ls='--',color='orange')
plt.show()
```

```python
n=nser_filtro.mean()
```

```python
def funcao_pop_0_14(populacao):
    pop_total=m*populacao+n
    return pop_total
```

```python
df['Populacao_Prevista_0_14']=df['População (0-14 anos)'].apply(funcao_pop_0_14)
```

```python
plt.scatter(df['População (0-14 anos)'],df['total_pop'],edgecolors='black')
plt.plot(df['População (0-14 anos)'],df['Populacao_Prevista_0_14'],ls='--')
plt.xlabel('População (0-14 anos)')
plt.ylabel('Pop. Total')
plt.show()
```
